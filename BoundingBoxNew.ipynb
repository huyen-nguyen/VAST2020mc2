```python
# ---------- IMPLEMENTATION - READ EVERY PATH ----------
import os
import csv
import cv2

def drawBox(row, img, line_count):
    def color(x):
        switcher = {
            0: (255, 0, 255), # fuchsia
            1: (0, 255, 0), # green
            2: (10, 255, 255), # neon green
            3: (0, 53, 255), # red
            4: (252, 255, 0), 
            5: (141, 135, 255), 

        }
        return switcher.get(int(x)%6)
    
    x = int(row["x"])
    y = int(row["y"])
    Width = int(row["Width"])
    Height = int(row["Height"])
    Label = row["Label"]
    Score = round(float(row["Score"]), 2)
                
    labelWidth = len(Label)*37 + 20
    
    # Rectangle ---------------------------------------
    recColor = color(line_count)
    cv2.rectangle(img, (x, y), (x+Width, y+Height), recColor, 8)
    
    cv2.rectangle(img, (x, y), (x + 200 if labelWidth < 200 else x + labelWidth, y + 140), recColor, -1)
    
    # score (2 digits) takes up 200px
        
    # TEXT ----------------------------------------

    # font 
    font = cv2.FONT_HERSHEY_DUPLEX

    # coordinates 
    orgLabel = (x + 5, y + 50) 
    orgScore = (x + 5, y + 120)

    # fontScale 
    fontScale = 2

    # Blue color in BGR 
    color = (0, 0, 0) 

    # Line thickness of 2 px 
    thickness = 2

    # Using cv2.putText() method 
    img = cv2.putText(img, f'{Label}', orgLabel, font,  
                       fontScale, color, thickness, cv2.LINE_AA) 

    img = cv2.putText(img, f'{Score}', orgScore, font,  
                       fontScale, color, thickness, cv2.LINE_AA) 

    
def drawAllBoxes(path):
    with open(path + ".csv", mode='r', encoding="latin-1") as csv_file:
        csv_reader = csv.DictReader(csv_file)
        line_count = 1

        img = cv2.imread(path + ".jpg", cv2.IMREAD_COLOR)

        for row in csv_reader:
#             print(f'box {line_count}: {row["Label"]}')
            if row["x"].isdigit() and row["y"].isdigit():
                drawBox(row, img, line_count)
                line_count += 1
            else:
                cv2.imwrite(path + "bbox.jpg",img)
                print("Error in data values at " + path)
                return 
            
        cv2.imwrite(path + "bbox.jpg",img)
        print("Done " + path)   # print once!!!
    


with os.scandir("MC2-Image-Data") as main_dir:
    for entry_dir in main_dir:
        if entry_dir.name.startswith("Person") and entry_dir.is_dir():
            with os.scandir(entry_dir.path) as person_dir:
                for entry in person_dir:
                    if entry.name.endswith(".csv") and entry.is_file():
                        path = entry.path[:-4]
                        print("Doing " + entry.name)
                        drawAllBoxes(path)
                
```
