```python
# ------- IMPLEMENTING CROP ON REAL IMAGE ----------

import os
import cv2
import csv

def cropBox(row, img, line_count):
    x = int(row["x"])
    y = int(row["y"])
    Width = int(row["Width"])
    Height = int(row["Height"])
    Label = row["Label"]
    
    crop_img = img[y:y+Height, x:x+Width].copy() 
    
    cv2.imwrite(path + "obj_" + str(line_count) + Label + ".jpg",crop_img)


def specialCrop(row, img, line_count):
    Label = row["Label"]
    cv2.imwrite(path + "obj_" + str(line_count) + Label + ".jpg", img)
    

def readPath(path):
    total_in_path = 0
    with open(path + ".csv", mode='r', encoding="latin-1") as csv_file:
        csv_reader = csv.DictReader(csv_file)
        
        img = cv2.imread(path + ".jpg", cv2.IMREAD_COLOR)
        
        line_count = 1
        
        for row in csv_reader:
            if row["x"].isdigit() and row["y"].isdigit():
                cropBox(row, img, line_count)
                line_count += 1
                total_in_path += 1
            else:
                print("Error in data values at " + path)
                specialCrop(row, img, line_count)
                line_count += 1
                total_in_path += 1
                
    return total_in_path
            

with os.scandir("MC2-Image-Data") as main_dir:
    total = 0
    for entry_dir in main_dir:
        if entry_dir.name.startswith("Person") and entry_dir.is_dir():
            with os.scandir(entry_dir.path) as person_dir:
                for entry in person_dir:
                    if entry.name.endswith(".csv") and entry.is_file():
                        path = entry.path[:-4]
                        print("Doing " + entry.name)
                        total += readPath(path)
               
    print(total)
```
